```python
import pandas as pd
import numpy as np

df = pd.read_csv('crime.csv', low_memory=False)
df.head()
```

```python
df.info()
```

```python
df.SHOOTING.fillna('N',inplace=True)
df.drop(['Location'],axis=1,inplace=True)
df.drop(['OCCURRED_ON_DATE'],axis=1,inplace=True)
df.drop(['STREET'],axis=1,inplace=True)
df.drop(['UCR_PART'],axis=1,inplace=True)
df.drop(['OFFENSE_DESCRIPTION','OFFENSE_CODE'],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
df.info()
```

### Fix extra space issue and drop repetitive data

### Replace object data types with integers

```python
for col in df.columns:
    if df[col].dtype == 'object':
        mapping = {}
        for i in range(1,len(df[col].unique())+1):
            mapping[df[col].unique()[i-1]] = i
        df[col].replace(mapping, inplace=True)

df.info()
```
