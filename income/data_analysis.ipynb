```python
import pandas as pd
import numpy as np

df = pd.read_csv('income_evaluation.csv')
df.head()
```

```python
df.info()
```

### Fix extra space issue and drop repetitive data

```python
df['workclass'] = df[' workclass']
df['fnlwgt'] = df[' fnlwgt']
df['education-num'] = df[' education-num']
df['marital-status'] = df[' marital-status']
df['occupation'] = df[' occupation']
df['relationship'] = df[' relationship']
df['race'] = df[' race']
df['sex'] = df[' sex']
df['capital-gain'] = df[' capital-gain']
df['capital-loss'] = df[' capital-loss']
df['hours-per-week'] = df[' hours-per-week']
df['native-country'] = df[' native-country']
df['income'] = df[' income']


df = df.drop([' workclass',' fnlwgt',' education',' education-num',
        ' marital-status',' occupation',' relationship',' race',
        ' sex',' capital-gain',' capital-loss',' hours-per-week',
        ' native-country',' income'],axis=1)
df.info()
```

### Replace object data types with integers

```python
for col in df.columns:
    if df[col].dtype == 'object':
        mapping = {}
        for i in range(1,len(df[col].unique())+1):
            mapping[df[col].unique()[i-1]] = i
        df[col].replace(mapping, inplace=True)

df.info()
```
