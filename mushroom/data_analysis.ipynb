```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mushrooms.csv')
df.head()
```

#### All the characterstics are stored as chars, for the sake of ease, it's better to make them ints

```python
class_mapping = {'p':0, 'e':1}
cap_shape_mapping = {'x':1,'b':2,'s':3,'f':4,'c':5,'k':6}
cap_surface_mapping = {'s':1,'y':2,'f':3,'g':4}
cap_color_mapping = {'n':1,'y':2,'w':3,'g':4,'e':5,'p':6,'b':7,'u':8,'c':9,'r':10}
bruise_mapping = {'t':1,'f':0}
odor_mapping = {'p':1,'a':2,'l':3,'n':4,'f':5,'c':6,'y':7,'s':8,'m':9}
gill_attach_mapping = {'f':1,'a':2}
gill_space_mapping = {'c':1,'w':2}
gill_size_mapping = {'n':1,'b':2}
gill_color_mapping = {}
stalk_shape_mapping = {}
stalk_root_mapping = {}
stalk_surf_a_ring_mapping = {}
stalk_surf_b_ring_mapping = {}
stalk_col_a_ring_mapping = {}
stalk_col_b_ring_mapping = {}
veil_type_mapping = {}
veil_color_mapping = {}
ring_num_mapping = {}
ring_type_mapping = {}
spore_color_mapping = {}
population_mapping = {}
habitat_mapping = {}
```

```python
for i in range(1,len(df['gill-color'].unique())+1):
    gill_color_mapping[df['gill-color'].unique()[i-1]] = i
for i in range(1,len(df['stalk-shape'].unique())+1):
    stalk_shape_mapping[df['stalk-shape'].unique()[i-1]] = i
for i in range(1,len(df['stalk-root'].unique())+1):
    stalk_root_mapping[df['stalk-root'].unique()[i-1]] = i
for i in range(1,len(df['stalk-surface-above-ring'].unique())+1):
    stalk_surf_a_ring_mapping[df['stalk-surface-above-ring'].unique()[i-1]] = i
for i in range(1,len(df['stalk-surface-below-ring'].unique())+1):
    stalk_surf_b_ring_mapping[df['stalk-surface-below-ring'].unique()[i-1]] = i
for i in range(1,len(df['stalk-color-above-ring'].unique())+1):
    stalk_col_a_ring_mapping[df['stalk-color-above-ring'].unique()[i-1]] = i
for i in range(1,len(df['stalk-color-below-ring'].unique())+1):
    stalk_col_b_ring_mapping[df['stalk-color-below-ring'].unique()[i-1]] = i
for i in range(1,len(df['veil-type'].unique())+1):
    veil_type_mapping[df['veil-type'].unique()[i-1]] = i
for i in range(1,len(df['veil-color'].unique())+1):
    veil_color_mapping[df['veil-color'].unique()[i-1]] = i  
for i in range(1,len(df['ring-number'].unique())+1):
    ring_num_mapping[df['ring-number'].unique()[i-1]] = i
for i in range(1,len(df['ring-type'].unique())+1):
    ring_type_mapping[df['ring-type'].unique()[i-1]] = i
for i in range(1,len(df['spore-print-color'].unique())+1):
    spore_color_mapping[df['spore-print-color'].unique()[i-1]] = i
for i in range(1,len(df['population'].unique())+1):
    population_mapping[df['population'].unique()[i-1]] = i
for i in range(1,len(df['habitat'].unique())+1):
    habitat_mapping[df['habitat'].unique()[i-1]] = i
```

```python
df['class'].replace(class_mapping,inplace=True)
df['cap-shape'].replace(cap_shape_mapping,inplace=True)
df['cap-surface'].replace(cap_surface_mapping,inplace=True)
df['cap-color'].replace(cap_color_mapping,inplace=True)
df['bruises'].replace(bruise_mapping,inplace=True)
df['odor'].replace(odor_mapping,inplace=True)
df['gill-attachment'].replace(gill_attach_mapping,inplace=True)
df['gill-spacing'].replace(gill_space_mapping,inplace=True)
df['gill-size'].replace(gill_size_mapping,inplace=True)
df['gill-color'].replace(gill_color_mapping,inplace=True)
df['stalk-shape'].replace(stalk_shape_mapping,inplace=True)
df['stalk-root'].replace(stalk_root_mapping,inplace=True)
df['stalk-surface-above-ring'].replace(stalk_surf_a_ring_mapping,inplace=True)
df['stalk-surface-below-ring'].replace(stalk_surf_b_ring_mapping,inplace=True)
df['stalk-color-above-ring'].replace(stalk_col_a_ring_mapping,inplace=True)
df['stalk-color-below-ring'].replace(stalk_col_b_ring_mapping,inplace=True)
df['veil-type'].replace(veil_type_mapping,inplace=True)
df['veil-color'].replace(veil_color_mapping,inplace=True)
df['ring-number'].replace(ring_num_mapping,inplace=True)
df['ring-type'].replace(ring_type_mapping,inplace=True)
df['spore-print-color'].replace(spore_color_mapping,inplace=True)
df['population'].replace(population_mapping,inplace=True)
df['habitat'].replace(habitat_mapping,inplace=True)
```
