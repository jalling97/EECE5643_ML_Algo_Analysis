Plotting the CPU utilization and runtime analysis from each dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from time import time
import psutil
import warnings
import random
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
phone_train_cpu = pd.read_csv('phone/phone_cpu.csv').values.flatten()
mushroom_train_cpu = pd.read_csv('mushroom/mushroom_cpu.csv').values.flatten()
income_train_cpu = pd.read_csv('income/income_cpu.csv').values.flatten()
crime_train_cpu = pd.read_csv('crime/crime_cpu.csv').values.flatten()

phone_runtime = pd.read_csv('phone/phone_runtime.csv').values.flatten()
mushroom_runtime = pd.read_csv('mushroom/mushroom_runtime.csv').values.flatten()
income_runtime = pd.read_csv('income/income_runtime.csv').values.flatten()
crime_runtime = pd.read_csv('crime/crime_runtime.csv').values.flatten()

phone_traintime = pd.read_csv('phone/phone_train_times.csv').values.flatten()
mushroom_traintime = pd.read_csv('mushroom/mushroom_train_times.csv').values.flatten()
income_traintime = pd.read_csv('income/income_train_times.csv').values.flatten()
crime_traintime = pd.read_csv('crime/crime_train_times.csv').values.flatten()

phone_acc = pd.read_csv('phone/phone_accuracy.csv').values.flatten()
mushroom_acc = pd.read_csv('mushroom/mushroom_accuracy.csv').values.flatten()
income_acc = pd.read_csv('income/income_accuracy.csv').values.flatten()
crime_acc = pd.read_csv('crime/crime_accuracy.csv').values.flatten()

phone_mem = pd.read_csv('phone/phone_peak_mem.csv').values.flatten() / 10**3
mushroom_mem = pd.read_csv('mushroom/mushroom_peak_mem.csv').values.flatten() / 10**3
income_mem = pd.read_csv('income/income_peak_mem.csv').values.flatten() / 10**3
crime_mem = pd.read_csv('crime/crime_peak_mem.csv').values.flatten() / 10**3
```

```python
# using example from https://python-graph-gallery.com/11-grouped-barplot/
plt.figure(figsize=(8,5))

# set width of bar
barWidth = 0.20
 
# Set position of bar on X axis
r1 = np.arange(len(phone_train_cpu))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
# Make the plot
plt.bar(r1, phone_train_cpu,    width=barWidth, edgecolor='white', label='phone dataset')
plt.bar(r2, mushroom_train_cpu, width=barWidth, edgecolor='white', label='mushroom dataset')
plt.bar(r3, income_train_cpu,   width=barWidth, edgecolor='white', label='income dataset')
plt.bar(r4, crime_train_cpu,    width=barWidth, edgecolor='white', label='crime dataset')
 
# Add xticks on the middle of the group bars
plt.xlabel('Algorithm', fontweight='bold')
plt.ylabel('CPU Utilization (%)', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(phone_cpu))], ['Nominal', 'KNN', 'Logistic Regression', 'SVM', 'Random Forest', 'MLP'])
 
# Create legend & Show graphic
plt.legend()
plt.title('Model Training CPU Utilization')
plt.savefig('train_cpu_utilization_all_data.png')
plt.show()
```
